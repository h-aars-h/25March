```python
## Q9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does
# the dataset have?
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
df1=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df1.shape
```

```python
char_remove_list=['+',"$",',']
col_remove=["Price","Installs"]
for item in char_remove_list:
    for col in col_remove:
        df1[col]=df1[col].str.replace(item,'')
```

```python
df1.info()
# 10841 rows and  13 columns 
```

```python
# Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
# app categories.
df1['Category'].value_counts().plot.pie(y=df1['Category'],figsize=(15,15),autopct='')
```

```python
import matplotlib.pyplot as plt 
# df1['Rating'].mean()
# df1['Rating'] = df1['Rating'].replace(np.nan,4.193338315362443)
numeric_features = [feature for feature in df1.columns if df1[feature].dtype != 'O']
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df1[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
# the relationship.
import matplotlib.pyplot as plt

plt.scatter(df1['Size'], df1['Rating'])
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()
```

```python
# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df1[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```
